```python
import sympy as sp
import random
import statistics
```

```python
def pdf(t, x):
    return (t+1)*((1-x)**t)* (sp.Heaviside(x) - sp.Heaviside(x-1))
def cdf_inverse(F):
    return (1-((1-F)**(1./3)))
```

q1_a

```python
x = sp.Symbol('x')
t = sp.Symbol('t')
cdf=sp.integrate(pdf(t,x), x)
t_value=input("please enter teta: ")
cdf_function = cdf.subs(t,t_value)
sp.plot(cdf_function,(x,-1,3))
```

q1_b

```python
prob=cdf.subs(t,2).subs(x,0.25)
print("pr(x<=0.25 | t=2) = ",prob)
```

q1_c

```python
TheList=[]
random.seed(80)
for i in range(0, 100000):
    k = random.random()
    value=float(cdf_inverse(k))
    TheList.append(value)
mean = statistics.mean(TheList)
variance = statistics.variance(TheList)
print("mean = ", mean)
print("variance = ", variance)
```
